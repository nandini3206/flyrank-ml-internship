## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
from IPython.display import Markdown, display

display(Markdown("""
# Why ML Beats a Fixed Rule

A fixed rule such as "refresh every page with low CTR" is too simple because webpage quality depends on many factors at the same time.

Machine learning can consider multiple signals together, such as search volume, CTR, impressions, word count, content type, competition, and engagement. This makes the recommendations more flexible and useful than relying on a single if-statement.
"""))
```
