## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

```python
from IPython.display import Markdown, display

display(Markdown("""
# Data Limits

This dataset has several important limitations that should be considered during analysis.

- The dataset contains anonymized client and content identifiers, so individual clients or websites cannot be identified.
- Client history is unbalanced, meaning different clients have different amounts of historical data.
- Some early records may contain only Google Search Console (GSC) information or only Google Analytics 4 (GA4) information.
- The dataset represents observational search performance data and cannot establish causal relationships.
- Future labels must be created carefully to avoid data leakage during model training.
"""))
```
