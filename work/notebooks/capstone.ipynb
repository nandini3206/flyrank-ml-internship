## 3. Methodology

*Assumptions, features, label definition, baseline, validation design, leakage checks.*

## Methodology

The objective of this project is to predict which content pages should be refreshed based on historical performance signals.

The dataset was cleaned before training by removing leakage-related features and handling missing values. Only features available before the prediction point were used so that the model reflects a realistic production scenario.

Several machine learning models were trained and evaluated. Their predictions were compared against a baseline scoring approach, and the model with the best validation performance was selected for generating the final refresh queue.

The resulting recommendations are intended to support human decision-making rather than automate content updates.

## 5. Limitations

*What this work cannot claim.*

## Limitations

This work demonstrates that machine learning can prioritize content for review, but it does not guarantee that refreshing content will improve search performance.

The model was trained on historical anonymized data, so performance may change if user behavior, search engine algorithms, or website structure changes.

The recommendations should always be reviewed by human editors before any content is updated. The model is intended as a decision-support tool rather than an automated publishing system.

Future work may include retraining with newer data, evaluating additional features, and monitoring model performance over time.

```python
print("Limitations")
print("- Historical data may not represent future behavior.")
print("- Human review is required before action.")
print("- Model performance depends on data quality.")
print("- Periodic retraining is recommended.")
```
